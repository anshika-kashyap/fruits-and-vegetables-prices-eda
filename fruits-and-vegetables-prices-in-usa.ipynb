# Introduction

This script focuses on executing Exploratory Data Analysis and creating Data Visualizations. In this script, we will use a high-quality Kaggle dataset.

# Importing Libraries

```r
# loading libraries
library(tidyverse)
library(dplyr)
library(ggplot2)

# checking if all the libraries imported successfully or not
print("Imported Successfully!")
```

```r
# loading the dataset
prices <- read_csv("/kaggle/input/fruits-and-vegetables-prices-in-usa/Prices.csv", show_col_types=FALSE)

# viewing the dataset
View(head(prices))
```

# Exploratory Data Analysis

```r
# renaming the columns to make them easier to read
colnames(prices) <- c("item","form","retail_price","retail_price_unit","yield","cup_equivalent_size","cup_equivalent_unit","cup_equivalent_price")

# rounding off the decimal places to 2 places
prices[, c("retail_price", "yield", "cup_equivalent_size", "cup_equivalent_price")] <- round(prices[, c("retail_price", "yield", "cup_equivalent_size", "cup_equivalent_price")], 2)

# viewing the dataset
View(head(prices))
```

```r
# checking for duplicate rows and null values

if (any(duplicated(prices))) {
  print("There are duplicate rows in the data frame.")
} else {
  print("There are no duplicate rows in the data frame.")
}

if (any(is.na(prices))) {
  print("There are empty cells in the data frame.")
} else {
  print("There are no empty cells in the data frame.")
}
```

# Data Visualization

## Visualization Creation

```r
# creating a horizontal bar graph
price <- prices %>% filter(retail_price >= 5)

ggplot(data = price)+
geom_bar(mapping = aes(x = retail_price, y = item), stat = "identity", fill = "tomato3")+
xlab("Retail Price")+
ylab("Item")+
labs(title = "Average Retail Price of Each Fruit and Vegetable
Which Had Retail Price Higher Than Or Equal To $5 in the Year 2020")
```

## Conclusion

**Data Insight**: Mangoes had the highest retail price in the year 2020.
